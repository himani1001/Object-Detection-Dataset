```python
import zipfile
import os

uploaded_file_path = '/content/archive.zip'
extract_path = '/content'

with zipfile.ZipFile(uploaded_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset extracted to:", extract_path)
```

```python
for root, dirs, files in os.walk(extract_path):
    for file in files:
        print(os.path.join(root, file))
```

```python
import torch

print("CUDA available:", torch.cuda.is_available())
print("GPU Device Name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "None")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data=os.path.join(extract_path, '/content/aquarium_pretrain/data.yaml'),
    epochs=5,
    imgsz=640,
    batch=4,
    name='custom_aquarium_detection',
    project='custom_training_project',
    workers=2,
    device='cuda'
)


print("Training complete.")
```

```python
metrics = model.val()
print("Validation Metrics:", metrics)
```

```python
test_image_path = os.path.join(extract_path, 'aquarium_pretrain', 'test', 'images')

predictions = model.predict(source=test_image_path, save=True)

for result in predictions:
    result.show()

print("Predictions complete and saved.")
```
